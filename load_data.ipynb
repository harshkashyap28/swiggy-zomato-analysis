```python
import pandas as pd
from sqlalchemy import create_engine
```

```python

engine = create_engine('postgresql://postgres:your_password@localhost:5432/postgres')

df = pd.read_csv('data/swiggy_vs_zomato_3000.csv')
```

```python
# fix boolean columns
df['has_own_website'] = df['has_own_website'].map({'True': True, 'False': False})
df['has_own_app'] = df['has_own_app'].map({'True': True, 'False': False})
df['food_license_verified'] = df['food_license_verified'].map({'True': True, 'False': False})
```

```python
# loading restaurants table
restaurants_cols = [
    'restaurant_id', 'restaurant_name', 'city', 'locality',
    'restaurant_type', 'cuisines', 'distance_from_city_center_km',
    'opening_time', 'closing_time', 'days_operational',
    'avg_cost_per_person_inr', 'price_category', 'amenities',
    'amenities_count', 'has_own_website', 'has_own_app',
    'food_license_verified', 'listing_date', 'days_listed'
]
df[restaurants_cols].to_sql('restaurants', engine, if_exists='append', index=False)
```

```python
# loading platform_metrics table
platform_cols = [
    'restaurant_id', 'swiggy_rating', 'swiggy_total_reviews',
    'zomato_rating', 'zomato_total_reviews', 'average_rating_both_platforms',
    'swiggy_delivery_fee_inr', 'zomato_delivery_fee_inr',
    'swiggy_avg_delivery_time_minutes', 'zomato_avg_delivery_time_minutes',
    'swiggy_platform_commission_pct', 'zomato_platform_commission_pct',
    'swiggy_discount_frequency_pct', 'zomato_discount_frequency_pct',
    'swiggy_estimated_monthly_orders', 'zomato_estimated_monthly_orders',
    'swiggy_estimated_monthly_revenue_inr', 'zomato_estimated_monthly_revenue_inr',
    'swiggy_estimated_net_profit_inr', 'zomato_estimated_net_profit_inr',
    'swiggy_market_share_pct', 'zomato_market_share_pct',
    'platform_performance_better'
]
df[platform_cols].to_sql('platform_metrics', engine, if_exists='append', index=False)

print("Data loaded successfully!")
```
